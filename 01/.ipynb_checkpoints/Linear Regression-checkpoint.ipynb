# Linear Regression

In general, linear models make prediction by simply computing a weighted sum of the input features, plus a constant called the bias term(also called the *intercept* term).

<pre>y = θ0 + θ1x1 + θ2x2 + ⋯ + θnxn</pre>

## Import libraries

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# print(matplotlib.colors.cnames)
```

```python
#print(X.shape, y.shape)
# Randomly generated data
fig = plt.figure(figsize=(10,8))
plt.scatter(X, y, s=30, c='steelblue', marker='o')
plt.xlabel("X")
plt.ylabel('y')
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)), X]
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

## Prediction

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
```

```python
y_predict
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(X_new, y_predict, 'r-')
plt.plot(X, y, 'b.')
plt.axis([0,2,0,15])
plt.show()
```
